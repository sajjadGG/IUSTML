DTs are extremely popular for a variety of reasons, being their interpretability probably their most important advantage. They can be trained very fast and are easy to understand, which opens their possibilities to frontiers far beyond scientific walls

### Pruning is a technique used to deal with overfitting , that reduces the size of DTs by removing sections of the Tree that provides little predictive or classification power

* Pre-prune: When you stop growing DT branches when information becomes unreliable.
* Post-prune: When you take a fully grown DT and then remove leaf nodes only if it results in a better model performance. This way, you stop removing nodes when no further improvements can be made.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

The iris data set contains four features, three classes of flowers, and 150 samples.
### Features: 
sepal length (cm), sepal width (cm), petal length (cm), petal width (cm)
### Classes :
setosa, versicolor, virginica --> [0 ,1 ,2]

```python
iris = load_iris()
X = pd.DataFrame(iris.data[:,:],columns = iris.feature_names[:])
y = pd.DataFrame(iris.target , columns=['Species'])

tree = DecisionTreeClassifier(max_depth = 2)
tree.fit(X,y)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(tree , 
               out_file = "myTreeName.dot",
               feature_names = list(X.columns),
               class_names = iris.target_names,
               filled = True,
               rounded = True)
```

```python
def gini(p):
    return 1 - p**2 -(1-p)**2
def entropy(p):
    return -p*np.log2(p) - (1-p)*np.log2(1-p)
def classificatin_error(p):
    return 1-np.max([p,1-p])

```
