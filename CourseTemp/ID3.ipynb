```python
def cal_entropy(sub_df , target='y' , plus=0 , minus=1):
    eps = np.finfo(float).eps
    num_pos=len(sub_df[sub_df[target]==plus])
    num_neg=len(sub_df[sub_df[target]==minus])
    total = num_pos + num_neg
    #always be caution around log when they are approaching zero
    return -1*(num_pos/total)*(np.log2(num_pos/total+eps)) -1*(num_neg/total)*(np.log2(num_neg/total+eps))
    
    
        
```

```python
def find_best_test(sub_df ,columns , target='y' ):
    #initialization
    min_ent = 2
    min_col=0
    total = len(sub_df)
    #print('total is {}'.format(total))
    b_groups = []
    for c in columns:
        groups=[]
        is_categorcal=True
        results=sub_df[c].unique()
        if(len(results) > 20):
            ## really sillllllly assumption
            is_categorical=False    
        sum_ent=0.0
        #print('-------{}----------'.format(c))
        if(is_categorical):
            for r in results:
                temp = sub_df[sub_df[c]==r]
                n=len(temp)
                ent=cal_entropy(temp,target)
                #print("columns is {} , result is {} ,len is {}, ent is {} ".format(c,r,n,ent))
                sum_ent += (n/total)*ent
                groups.append(temp)
        else:
            min_ent_sub = 2
            min_r_sub=0
            for i in range(len(sub_df)-1):
                sum_ent_sub=0.0
                sub_groups=[]
                r = (sub_df.iloc[i][c] + sub_df.iloc[i+1][c])/2
                tempr = sub_df[sub_df[c]>r]
                n=len(tempr)
                ent=cal_entropy(tempr,target)
                #print("columns is {} , result is {} ,len is {}, ent is {} ".format(c,r,n,ent))
                sum_ent_sub += (n/total)*ent
                templ = sub_df[sub_df[c]>r]
                n=len(templ)
                ent=cal_entropy(templ,target)
                #print("columns is {} , result is {} ,len is {}, ent is {} ".format(c,r,n,ent))
                sum_ent_sub += (n/total)*ent
                sub_groups.append(tempr)
                sub_groups.append(templ)
                if(sum_ent_sub<min_ent_sub):
                    min_ent_sub = sum_ent_sub
                    min_r_sub = r
                    groups = sub_groups.copy()
            sum_ent = min_ent_sub
        #print("sum_ent is {}".format(sum_ent))
        if(sum_ent<min_ent):
            min_ent = sum_ent
            min_col = c
            b_groups = groups
    return min_col , min_ent ,groups
```

```python
def remove(l,c):
    l2=[]
    for m in l:
        if(m in c):
            continue
        else:
            l2.append(m)
    return l2
```
