```python
def getFeatures(df , WINDOW=5 , train_months=6 , test_months=1 , is_debug=False,y_label='close'):
    dftrain = df.head((train_months+test_months)*30)
    dftest  = dftrain.head(test_months*30)
    dftrain = dftrain.tail(train_months*30)
    
    #get features 
    
    WINDOW=10
    x_train=dftrain.iloc[0:WINDOW].values.flatten()
    y_train=np.array([dftrain.iloc[WINDOW][y_label]])
    for i in range(1,len(dftrain)-WINDOW):
        x_train=np.vstack([x_train,dftrain.iloc[i:i+WINDOW].values.flatten()])
        y_train=np.concatenate([y_train,[dftrain.iloc[i+WINDOW][y_label]]])
    if(is_debug):
        print("x_train is {} \n, shape is {}".format(x_train,x_train.shape))
        print("y_train is {} \n shape is {}".format(y_train,y_train.shape))
        
    #Test 
    
    WINDOW=10
    x_test=dftest.iloc[0:WINDOW].values.flatten()
    y_test=np.array([dftest.iloc[WINDOW][y_label]])
    for i in range(1,len(dftest)-WINDOW):
        x_test=np.vstack([x_test,dftest.iloc[i:i+WINDOW].values.flatten()])
        y_test=np.concatenate([y_test,[dftest.iloc[i+WINDOW][y_label]]])
    if(is_debug):
        print("x_test is {} \n, shape is {}".format(x_test,x_test.shape))
        print("y_test is {} \n shape is {}".format(y_test,y_test.shape))
    
    return x_train , y_train , x_test , y_test
    
```

```python
def datedif(d1 , d2):
    #print(d1,d2)
    if(type(d1)!=str):
        d1=str(d1).split()[0]
        d2=str(d2).split()[0]
    print(d1)
    d1 = tuple(map(int,d1.split('-')))
    d1 = d1[2] + d1[1] *30 + d1[0]*365
    d2 = tuple(map(int,d2.split('-')))
    d2 = d2[2] + d2[1] *30 + d2[0]*365
    return d2-d1
def gap(df):
    gap = np.zeros(len(df))
    gap[-1]=0
    for i in range(len(df)-1):
        gap[i] = datedif(df.iloc[i+1].name , df.iloc[i].name)
    return gap
```

```python
def clean_csv(filename):
    #date
    df = pd.read_csv(filename)
    df2 = pd.DataFrame({'year' : [str(x)[0:4] for x in df['<DTYYYYMMDD>']],
                   'month' : [str(x)[4:6] for x in df['<DTYYYYMMDD>']],
                   'day' : [str(x)[6:8] for x in df['<DTYYYYMMDD>']]})
    df['date'] = pd.to_datetime(df2)
    
    #names and dropping redundent columns
    df['first'] = df['<FIRST>']
    df['high'] = df['<HIGH>']
    df['low'] = df['<LOW>']
    df['close'] = df['<CLOSE>']
    df['value'] = df['<VALUE>']
    df['volume'] = df['<VOL>']
    df=df[['date' , 'first' , 'high' , 'low' , 'close' , 'value' , 'volume']]
    df.set_index('date',inplace=True)
    #df=df.drop(['Unnamed: 0'] , axis=1)
    df.sort_index(ascending=False)
    #print(df.head())
    #adding new features
    df['gap'] = gap(df)
    df['dif'] = (df['close'] - df['first'])/(df['high']+df['low'])
    return df
```

```python
def train_report(df,reg):
    x_train , y_train , x_test , y_test = getFeatures(df)
    reg.fit(x_train, y_train)
    print("score on train {} \n score on test {}".format(reg.score(x_train, y_train) , reg.score(x_test, y_test)))
    xaxis = [i for i in range(x_test.shape[0])]
    plt.plot(xaxis,y_test)
    plt.plot(xaxis , reg.predict(x_test),'r')
    plt.title("Test")
    plt.show
```

```python
def insert_fix(l,t , normalize=True , np=0.05):
    if abs((t-l[-1]))/l[-1] > np:
        t=(1+np)*l[-1]
    l=list(l)
    l2=l[1:]
    l2.append(t)
    return l2
```

```python
def custom_loss2(y_true , y_pred , M=100000):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    y_prev = np.concatenate([y_true[0]*np.ones(1),y_true[0:-1]])
    return np.sum((y_pred-y_true)**2+ M*vec_min(np.zeros(len(y_pred)),(abs((y_pred-y_prev)/y_prev)-0.05)))
```

```python
def custom_loss1(y_true , y_pred , M=100000):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    #print('y_true.shape : {}  , y_pred.shape : {}'.format(y_true.shape,y_pred.shape))
    y_prev = np.concatenate([y_true[0]*np.ones(1),y_true[0:-1]])
    #derivative ?
    return np.sum((y_pred-y_true)**2) + M*np.min([0,(abs((y_pred-y_prev)/y_prev)-0.05).any()])
```
